```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a)
a = -1
I = 5
T = 50

def solver(I, T, theta, dt):
    """Solve u'=-a*u, u(0)=I, for t in (0,T] with steps of dt."""
    dt = float(dt)            # avoid integer division
    Nt = int(round(T/dt))     # no of time intervals
    T = Nt*dt                 # adjust T to fit time step dt
    u = np.zeros(Nt+1)           # array of u[n] values
    t = np.linspace(0, T, Nt+1)  # time mesh

    A = (1 - (1 - theta) * a * dt) / (1 + theta * a * dt)

    u[0] = I                  # assign initial condition
    for n in range(0, Nt):    # n=0,1,...,Nt-1
        u[n+1] = A*u[n]
    return u, t
```

```python
def u_exact(a, I, t):
    return I*np.exp(-a*t)
```

```python
t = np.linspace(0, 50)
u_e = u_exact(a, I, t)
```

```python
theta0_dt0, t00 = solver(I, T, theta=0, dt=0.1)
theta0_dt1, t01 = solver(I, T, theta=0, dt=0.5)
theta0_dt2, t02 = solver(I, T, theta=0, dt=0.75)
theta0_dt3, t03 = solver(I, T, theta=0, dt=1.25)
```

```python
plt.plot(t00, theta0_dt0, label = "dt = 0.1")
plt.plot(t01, theta0_dt1, label = "dt = 0.5")
plt.plot(t02, theta0_dt2, label = "dt = 0.75")
plt.plot(t03, theta0_dt3, label = "dt = 1.25")
plt.plot(t, u_e, label = "exact")
plt.legend()
plt.title("Theta = 0")
```
